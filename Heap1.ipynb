# <b>Heap(힙) 이란?</b>

데이터에서 최대값과 최소값을 빠르게 찾기 위해 고안된 완전 이진 트리(Complete Binary Tree)


+ 완전 이진 트리 : 노드를 삽입할 때 최하단 왼쪽 노드부터 차례대로 삽입하는 트리

<b>왜 힙을 사용하는가? </b>

+ 배열에 데이터를 넣고, 최대값과 최소값을 찾으려면 O(n)이 걸림


+ 이에 반해 힙에 데이터를 넣고, 최대값과 최소값을 찾으면, O(logn)이 걸림


+ 우선순위 큐와 같이 최대값 또는 최소값을 빠르게 찾아야 하는 자료구조 및 알고리즘 구현 등에 활용됨.

# 힙(Heap)의 구조

+ 힙은 최대값을 구하기 위한 구조인 Max Heap, 최소값을 구하기 위한 구조인 Min Heap으로 분류할 수 있다.





+ <span style='font-size:15pt'> 힙은 다음과 같이 두 가지 조건을 가지고 있는 자료구조이다. </span>

    1. 각 노드의 값은 해당 노드의 자식 노드가 가진 값보다 크거나 같다(최대 힙의 경우) => 가장 위에 노드가 가장 큰 값을 가진다.
        
        - 최소 힙의 경우는 각 노드의 값은 해당 노드의 자식 노드가 가진 값보다 작거나 같음. => 가장 위의 노드가 가장 작은 값을 가진다.
    
    
    => 따라서 Max, min을 구하고 싶을 때 데이터를 다 탐색할 필요 없이 root node만 빼오면 되기 때문에 압도적으로 빠르다
    
     2. 완전 이진 트리 형태를 가짐.

## 완전 이진 트리 , 힙 삽입 구조

왼쪽 -> 오른쪽 -> 왼쪽 아래의 왼쪽 -> 왼쪽 아래의 오른쪽 -> 오른쪽 아래의 왼쪽 -> 오른쪽 아래의 오른쪽 ....

## 힙과 이진 탐색 트리의 공통점과 차이점

+ 공통점 : 힙과 이진 탐색 트리는 모두 이진 트리임


+ 차이점 : 
    
    
    - 힙은 각 노드의 값이 자식 노드보다 크거나 같음(Max Heap)
    
    - 이진 탐색 트리는 왼쪽 자식 노드의 값이 가장 작고, 그 다음 부모 노드, 그 다음 오른쪽 자식 노드 값이 가장 큼
    
    - 힙은 이진 탐색 트리의 조건인 자식 노드에서 작은 값은 왼쪽, 큰 값은 오른쪽이라는 조건은 없음
    
        - 힙의 왼쪽 및 오른쪽 자식 노드의 값은 오른쪽이 클 수도 있고, 왼쪽이 클 수도 있음
    
    
    
+ 이진 탐색 트리는 탐색을 위한 구조, 힙은 최대/최소값 검색을 위한 구조 중 하나로 이해한다.
