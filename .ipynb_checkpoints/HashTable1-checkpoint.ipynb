# 해쉬 구조

+ Hash Table : 키(key)에 데이터(value)를 저장하는 데이터 구조
    - Key를 통해 바로 데이터를 받아올 수 있기 때문에 속도가 매우 빠름.
    - Python에는 해쉬를 별도로 구현할 필요가 없다! 딕셔너리 타입을 사용하면 되기 때문.
    - 보통은 배열로 미리 해쉬 테이블 사이즈만큼 생성 후에 사용한다.

+ 해쉬(Hash) : 임의 값을 고정 길이로 변환하는 것. (얼마나 길던지 고정된 길이로 변환)


+ 해쉬 테이블 (Hash Table) : 키 값의 연산에 의해 직접 접근이 가능한 데이터 구조 (해시 주소, 값에 대한 정보를 가지고 있는 데이터 구조)


+ 해싱 함수 (Hashing function) : 키를 입력으로 받아서 특정 데이터의 위치를 리턴하는 함수 => 자기가 임의로 만들거나 이미 만들어진 함수를 사용하면 됨


+ 슬롯(Slot) : 한 개의 데이터를 저장할 수 있는 공간


+ <b>저장할 데이터에 대해 Key를 추출할 수 있는 별도 함수도 존재할 수 있다.</b>

# 간단한 해쉬 예

```python
#hash_table = list([i for i in range(10)])
hash_table = list([0 for i in range(10)])
hash_table #index가 해쉬 주소라고 생각할 수 있음.
```

## 해쉬 함수 예

```python
#다양한 해쉬 함수 고안 기법이 있다. 가장 간단한 방식인 Division 법 (나누기를 통한 나머지의 값을 사용하는 기법)
def hash_func(key):
    return key % 5 #5의 나머지이므로 고정된 길이의 주소로 나오게 됨.
```

```python
data1 = 'Andy'
data2 = 'Dave'
data3 = 'Trump'

#ord() : 문자의 ASCII(아스키) 코드로 리턴, unicode나오기 이전에 사용하던 방식. 단점은 영어를 제외한 문자는 전환불가능
print(ord(data1[0]), ord(data2[0]), ord(data3[0]))
print(ord(data1[0]), ord(data1[0]) % 5)
print(ord(data1[0]), hash_func(ord(data1[0]))) #ord(data1[0]) 이 Key 값이 되는 것임. 
```

## 해쉬 테이블에 값 저장 예

```python
#value : data에 실제로 저장하고 싶은 값
#data로 키 값을 만들고 키 값으로 해쉬 함수에 넣어서 주소를 얻고 이 주소에 해당하는 slot에 실제 저장하고 싶은 value를 집어넣음.
#dictionary 형태와 굉장히 유사해짐.
def storage_data(data, value):
    key = ord(data[0])
    hash_address = hash_func(key)
    hash_table[hash_address] = value
```

```python
#데이터 저장하기
storage_data('Andy','01055553333')
storage_data('Dave','01055554444')
storage_data('Trump','01055552222')
```

```python
#데이터 읽기
def get_data(data):
    key = ord(data[0])
    hash_address = hash_func(key)
    return hash_table[hash_address]
```

```python
get_data('Andy')
```
