<span style ="font-size:15pt"> Linked List의 복잡한 기능 : 특정 노드를 삭제 </span>

<경우의 수>
1. 헤드(맨 앞의 노드 A)를 삭제해야한다면, 헤드가 바뀌어야 함. 
2. 맨 마지막 노드를 없애려면 마지막 노드 전의 주소값을 Null로 바꿔줘야함
3. 중간 노드의 경우를 삭제할 때 : 앞의 주소값을 C로 바꿔줘야한다.

```python
#앞에서 구현했던 클래스에 delete 함수를 추가해주자.

class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
    
class NodeMgmt:
    def __init__(self, data):
        self.head = Node(data)
    
    def add(self, data):
        if self.head == '':
            self.head = Node(data)
        else:
            node = self.head
            while node.next:
                node = node.next
            node.next = Node(data)
    def desc(self): #describe(출력 함수)
        node = self.head
        while node:
            print(node.data)
            node = node.next
    
    def delete(self, data):
        #방어코드 : 헤드가 없는 경우를 대비
        if self.head == '':
            print("해당 값을 가진 노드가 없습니다.")
            return #그냥 끝내는 것
        
        #1. head 삭제의 경우
        if self.head.data == data: 
            temp = self.head #현재 헤드를 임시 변수에 할당. 왜냐하면 밑에서 객체를 삭제하려고.
            self.head = self.head.next #A의 주소값 부분이 B의 헤드로 설정되어야 함
            del temp #기존에 삭제하려던 객체 삭제
        
        #2, 3 번 케이스를 다 합함. 이거 경우따져가면서 이해하기 
        else:
            node = self.head #head는 데이터가 아님
            while node.next:
                if node.next.data == data:#다음 노드의 데이터가 삭제할 데이터냐고 물어보는 것
                    #node.next 가 현재 B에 위치해있다고 생각하면 됨.
                    #맞다면 아래 수행
                    temp = node.next
                    node.next = node.next.next
                    #node.next : A의 주소값 = B를 말함. 즉 이 값이 C를 나타내도록 해줌
                    #node.next.next가 C니까.
                    del temp
                    return
                else:
                    node = node.next #만약 내가 찾는 삭제할 노드가 아니면 다음 노드로 넘어감
                
```

2,3번 케이스 부분 그림으로 표현

중간의 노드를 없애는 경우

마지막 노드를 없애고 싶을 때 node와 node.next 지정법


<span style="font-size:13pt"> 테스트를 위해 코드 넣어보기 </span>

```python
linkedlist1 = NodeMgmt(0)
linkedlist1.desc()
```

```python
#linked list 의 head가 살아있음을 확인
linkedlist1.head
```

```python
#head 지우기
linkedlist1.delete(0)
```

```python
#다시 head 값 출력
linkedlist1.head #아무것도 나오지 않는다 : 해당 객체가 사라졌다.
```

```python
#다시 노드를 만들기
linkedlist1 = NodeMgmt(0)
linkedlist1.desc()
```

```python
#여러 개의 노드를 더 추가하기
for data in range(1, 10):
    linkedlist1.add(data)
linkedlist1.desc()
```

```python
#중간 데이터 삭제
linkedlist1.delete(4)
```

```python
linkedlist1.desc() #4가 삭제되었음을 확인할 수 있다.
```

```python
#마지막 노드 삭제
linkedlist1.delete(9)
```

```python
linkedlist1.desc() #맨 뒤의 9가 삭제되었음을 알 수 있다.
```
