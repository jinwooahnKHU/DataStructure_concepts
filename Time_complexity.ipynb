# 알고리즘 복잡도 계산이 필요한 이유

<b>하나의 문제를 푸는 알고리즘은 다양할 수 있음</b>

ex. 정수의 절대값 구하기
+ 정수값을 제곱한 값에 다시 루트 씌우기
+ 정수가 음수인지 확인하고 음수일 때만 -1 곱하기



다양한 알고리즘 중 어느 알고리즘이 더 좋은지를 분석하기 위해, 복잡도를 정의하고 계산한다.

# 알고리즘 복잡도 계산 항목

1. 시간 복잡도 : 알고리즘 실행 속도
    + 반복문을 가지고 계산한다
    
    
2. 공간 복잡도 : 알고리즘이 사용하는 메모리 사이즈

if문이 몇개인지, 변수를 몇개 선언했는지가 아니라 반복문이 얼마나 오래 걸리느냐가 핵심!

# 알고리즘 성능 표기법

<b>Big O(빅-오 표기법) : O(N) </b>
    + 알고리즘 최악의 실행 시간을 표기 -> 아무리 최악의 상황이라도 이정도의 성능은 보장한다는 뜻.
    + 가장 많이/일반적으로 사용

<b>Ω(오메가) 표기법 : Ω(N) </b>
    + 알고리즘 최상의 실행 시간을 표기

<b>⍬(세타) 표기법 : ⍬(N)</b>
    + 세타 표기법은 알고리즘 평균 실행 시간을 표기


# 내 알고리즘의 빅 O 표현법

<span style='font-size:20pt; color:red;'> 단순하게 입력 n에 따라, 몇 번 실행이 되는지를 계산하면 된다!!</span>


+ 표현식에 가장 큰 영향을 미치는 n(입력)의 단위로 표기한다.
    
+ n이 1이든 100이든 1000이든 실행을 

    + 무조건 2회(상수회) 실행한다 : O(1)

    ``` python
            if n> 10:
                print(n)
    ```
    
    
     
+  n에 따라 n번, n+10번, 3n+10번 실행한다 : O(n)    
    + n에 따라서 n번 실행 하는 예시

``` python
    variable = 1
    for index in range(n):
        print(index)
```
        n이 100번이면 반복문 100번 + variable 선언 1 = 101번 실행


    + 3n + 1 예시
    
``` python
    variable = 1
    for num in range(3):
        for index in range(n):
            print(index)
```



+ n에 따라 n^2 번, n^2 + 1000번 등 실행한다 : O(n^2)

``` python
    variable = 1
    for num in range(n):
        for index in range(n):
            print(index)
```

# 실제 알고리즘을 예로 시간 복잡도 및 빅 오 표기법 알아보기

## 1~ n까지의 합을 구하는 알고리즘

````python
#algorithm 1
def sum_all(n):
    total = 0
    for num in range(1,n+1):
        total += num
    return total
````

````python
#algorithm 2
def sum_all(n):
    return int(n *(n+1)/2)
````

````python
sum_all(100)
````

Algorithm 1 : n번 실행 => O(n)

Algorithm 2 : 1번 실행 => O(1)
